```python
def get_con_describe(DF, cols_con):
    DF_describe_con = DF[cols_con].describe().T
    missing = 100 * (DF[cols_con].isnull().sum()/DF.shape[0])
    card = DF[cols_con].nunique()
    DF_describe_con = pd.concat([DF_describe_con, pd.DataFrame(missing, columns=['%missing']), pd.DataFrame(card, columns=['card'])], axis=1)
    return DF_describe_con
```

## categorical features statistics

```python
cols_cat = ["WEEK_DAY","PERIOD","condition", "STARTEND"]
```

```python
def generate_mode(DF, cols):
    """
    For categorical data
    Input: pd.DataFrame with specific cols
    Return: pd.DataFrame of Mode, Mode freq., Mode %, 2nd Mode, 2nd Mode freq., 2nd Mode % cols
    """
    # init Series
    mode_1st = pd.Series(dtype="object")
    mode_1st_freq = pd.Series(dtype="float64")
    mode_1st_rate = pd.Series(dtype="float64")

    mode_2nd = pd.Series(dtype="object")
    mode_2nd_freq = pd.Series(dtype="float64")
    mode_2nd_rate = pd.Series(dtype="float64")

    for col in cols:
        ans = DF[col].value_counts()
        # get 1st
        mode_1st = mode_1st.append(pd.Series(ans.nlargest(2).index[0], index=[col]))
        mode_1st_freq = mode_1st_freq.append(pd.Series(ans.nlargest(2)[0], index=[col]))
        mode_1st_rate = mode_1st_rate.append(pd.Series(100 * (mode_1st_freq / DF[col].count()), index=[col]))

        mode_2nd = mode_2nd.append(pd.Series(ans.nlargest(2).index[1], index=[col]))
        mode_2nd_freq = mode_2nd_freq.append(pd.Series(ans.nlargest(2)[1], index=[col]))
        mode_2nd_rate = mode_2nd_rate.append(pd.Series(100 * (mode_2nd_freq / DF[col].count()), index=[col]))
    return pd.DataFrame({'Mode':mode_1st,'Mode Freq.':mode_1st_freq,'Mode %':mode_1st_rate,
                      '2nd Mode':mode_2nd ,'2nd Mode Freq.': mode_2nd_freq, '2nd Mode %': mode_2nd_rate})
```

```python
DF_describe_cot = DF_filtered[cols_cat].describe().T.drop(['unique', 'top', 'freq'], axis=1)

card = pd.DataFrame(DF_filtered[cols_cat].nunique(), columns=['card'])
missing = pd.DataFrame(100 * (DF_filtered[cols_cat].isnull().sum()/DF_filtered.shape[0]), columns=['%missing'])
modes = generate_mode(DF_filtered, cols_cat)
DF_describe_cot = pd.concat([DF_describe_cot, card, missing, modes], axis=1)

del(missing)
del(card)
del(modes)
DF_describe_cot
```

## Reduce card of `condition`

```python
DF_filtered["condition"].value_counts().index
```

```python
def condition_reduce(x):
    bad = ['Light Rain','Light Drizzle','Light Rain / Windy', 'Light Drizzle / Windy','Light Rain Shower / Windy','Light Snow Shower',
           'Light Snow Shower / Windy', 'Rain','Mist', 'Fog','Shallow Fog', 'Wintry Mix / Windy',  'Light Snow / Windy','Rain Shower / Windy',
           'Rain / Windy','Drizzle','Wintry Mix','Patches of Fog','Heavy Rain', 'Light Snow' ,'Light Rain Shower','Showers in the Vicinity']
    good = ['Mostly Cloudy', 'Fair', 'Partly Cloudy', 'Mostly Cloudy / Windy', 'Partly Cloudy / Windy', 'Fair / Windy', 'Cloudy','Cloudy / Windy',]
    if x in  good:
        return 'Fair'
    else:
        return 'Bad'
```

```python
DF_filtered["condition"] = DF_filtered["condition"].apply(condition_reduce)
```
